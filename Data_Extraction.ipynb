```python
#import
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
```

```python
#Setting Up driver

#driver path that I am using
driverPath = "C:\Users\Sun of sadi\Data Extraction from firefox\geckodriver-v0.29.1-win64\geckodriver.exe"

#sometimes webdriver cannot find the firefox exe file at its designated place. So I am providing its path to the webdriver using options
options = Options()
options.binary_location = r"C:\Program Files\Mozilla Firefox\firefox.exe"
driver = webdriver.Firefox(options=options, executable_path=driverPath)
```

```python
#Getting the webPage and extracting all the categories


#URL of the desired page
url= 'https://addons.mozilla.org/en-CA/firefox/extensions'

driver.get(url)
time.sleep(5)

#Parsing into html page using html.parser
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

#finding all categories and their respective links
categories_html = soup.find_all('li', {"class": "Categories-item"})
categories = [cat.a.text for cat in categories_html]

#adding the website name with results to get the exact address of pages
link_cats = ['https://addons.mozilla.org'+cat.a.get('href') for cat in categories_html]
```

```python
categories
```

```python
#variable to store all data
extension_name = []
summary = []
users = []
author = []
rating = []
category = []
link = []
screenshot = []
num_of_release = []
permissions = []
info = []
```

```python
#function to ectraxt data from page using category num
def extract_page(num,page,link_page_1):
    next_link = link_page_1
    if page != 1:
        next_link = next_link+"?page="+str(page)
    driver.get(next_link)
    time.sleep(2)
    html_ = driver.page_source
    soup_ = BeautifulSoup(html_, 'html.parser')

    names = soup_.find_all('a', {'class': 'SearchResult-link'})
    summaries = soup_.find_all('p', {'class': 'SearchResult-summary'})
    usertxt = soup_.find_all('span', {'class': 'SearchResult-users-text'})
    authors = soup_.find_all('h3', {'class': 'SearchResult-author'})
    ratings = soup_.find_all('div', {'class': 'Rating'})
    plugin_links = soup_.find_all('a', {'class': 'SearchResult-link'})
    link_plug = ['https://addons.mozilla.org' + link.get('href') for link in plugin_links]
        
        #print(link_plug)


    #getting data and storing them inside global list
    for k in range(0, len(names)):
            #storing name in list
        extension_name.append(names[k].text)

            #storing summary in list
        summary.append(summaries[k].text)

            #storing developer or author in list
        author.append(authors[k].text)

            #storing ratings in list
        rating.append(ratings[k].get('title'))

            #storing category in list
        category.append(categories[num])

            #storing link in list
        link.append(link_plug[k])

            #extracting users
        txt = usertxt[k].text
        nums_ = [word_ for word_ in txt.split(" ")]
            #storing user in list
        users.append(nums_[0])

            # getting inside plugin page to extract version info
        driver.get(link_plug[k])
        time.sleep(3)
        html__ = driver.page_source
        soup__ = BeautifulSoup(html__, 'html.parser')
            
            #checking weather there is any data or not
        msg = soup__.find('header',{'class':'Card-header'})
        msg = msg.text
        if msg == 'Oops! We canâ€™t find that page':
            users.pop()
            link.pop()
            category.pop()
            rating.pop()
            author.pop()
            summary.pop()
            extension_name.pop()
            continue
        
            #extracting screenshots
        scrnshots = soup__.find_all('img',{'class':'ScreenShots-image'})
        scrnshots = [shot.get('src') for shot in scrnshots]
        strShot = ' ,'.join(scrnshots)
            #storing screenshots in list
        screenshot.append(strShot)

            #extracting permissions
        permissions_ = soup__.find_all('span',{'class':'Permission-description'})
        permissions_ = [per.text for per in permissions_]
        permissions_ = " ,".join(permissions_)
            #storing permission in list
        permissions.append(permissions_)


            #getting inside version info page of each plugin
        link_version ="https://addons.mozilla.org"+ soup__.find('a',{'class':'AddonMoreInfo-version-history-link'}).get('href')
        driver.get(link_version)
        time.sleep(2)
        html___ = driver.page_source
        soup___ = BeautifulSoup(html___, 'html.parser')
            #extracting all the info we need about the plugin
        versions_ = soup___.find_all('h2',{'class':'AddonVersionCard-version'})
        versions_ = [version.text for version in versions_]
        versions_ = [version.split()[1] for version in versions_]

            #storing num of release in list
        num_of_release.append(len(versions_))
            #print(versions_)
            #extracting version info
        releases_ = soup___.find_all('div',{'class':'AddonVersionCard-fileInfo'})
        releases_ = [release.text for release in releases_]
        releases_date = [release.split('-')[0] for release in releases_]
        releases_date = [release.split('Released')[1] for release in releases_date]
        file_size = [release.split('-')[1] for release in releases_]
        compatibility = soup___.find_all('div', {'class':'AddonVersionCard-compatibility'})
        compatibility = [com.text for com in compatibility]
        compatibility = [com.split('Works with')[1] for com in compatibility]
        Infos = []
        for o in range(0,len(file_size)):
            size = file_size[o]
            date = releases_date[o]
            version = versions_[o]
            compat = compatibility[o]
            val = {'version':version, 'date' : date, 'size' : size, 'compatibility' : compat}
                #Infos.append(VersionInfo(version=version, date= date, size=size))
            Infos.append(val)
            #storing version info in list
        info.append(Infos)


        # returing to catgory page
    driver.get(next_link)
    time.sleep(3)
    html_ = driver.page_source
    soup_ = BeautifulSoup(html_, 'html.parser')
```

```python
#function to ectraxt data from a category

def extract_category(num=0):
    #extracting process
    driver.get(link_cats[num])
    time.sleep(3)
    html_ = driver.page_source
    soup_ = BeautifulSoup(html_, 'html.parser')

    # finding the num of pages any category have
    page_nums = soup_.find('div', {'class': 'Paginate-page-number'})
    page_str = page_nums.text
    nums = [int(word) for word in page_str.split() if word.isdigit()]
    pages = nums[1]+1
    next_link = link_cats[num]

    # collecting data from all pages for every categories
    for j in range(1, pages):
        extract_page(num, j,next_link)
```

```python
#using loop to get the rest of the data
for i in range(0,len(categories)):
    extract_category(i)
    
    #for safety of data, saving everytime after finishing each category
    df = pd.DataFrame(
    {'Product Name': extension_name, 'Summary': summary, 'Developer': author, 'Users': users,
     'Rating': rating,'Link' : link,'ScreenShot':screenshot, 'Category': category,'Num Of Versions': num_of_release,'Permissions': permissions, 'Version-Info' : info})
    df.to_csv("final_data.csv")
    
```

```python
#Let's convert ratings into nums
def extract_rating(x):
    if x == 'There are no ratings yet' :
        return 0
    else:
        ex = [a for a in x.split()]
        num = float(ex[1])
        return num
```

```python
data['Rating'] = data['Rating'].apply(extract_rating)
```
